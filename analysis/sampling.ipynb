```python
import pandas as pd
df = pd.read_csv('cleaned_final.csv')
df
```

```python
df['period'] = None
df[(df['date'] < '2020-02-24') & (df['date'] > '2020-01-31')]
```

```python
# assigning priod number to different dates
df.loc[df['date'] < '2020-02-24', 'period'] = 0
df.loc[(df['date'] < '2020-03-24') & (df['date'] > '2020-02-23'), 'period'] = 1
df.loc[(df['date'] < '2020-05-04') & (df['date'] > '2020-03-23'), 'period'] = 2
df.loc[df['date'] > '2020-05-03', 'period'] = 3
```

```python
# list of percentages of full df in each period
fracs = []
for i in df.groupby(['period'])['period'].count():
    fracs.append(i / 4600)
fracs
```

```python
sample_size = 200
sample = pd.DataFrame()
for i in range(4):
    sample = sample.append(df.loc[df['period'] == i].sample(int(fracs[i]*sample_size)))
sample
```
