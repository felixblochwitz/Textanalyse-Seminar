## Schritt 1: Identifikation fehlerhaft importierter Artikel anhand der Publikation

```python
data = pd.read_csv('final.csv')
```

```python
data
```

```python
# checking for null values in publication column
data[data.publication.notnull()]
```

```python
data.publication.unique()
```

```python
pd.options.display.max_rows = 250
counts = data.publication.value_counts()
counts_df = pd.DataFrame({'publication':counts.index, 'count':counts.values})
counts_df
```

```python
to_drop = counts_df.loc[counts_df['count'] == 1]
to_drop
# 238 'Quellen' die nur einmal erscheinen
```

```python
# Löschen der Zeilen aus dem Datensatz in denen sich fehlerhaft geparsete Publikationen befinden
data = data[~data.publication.isin(to_drop.publication)]
# Wiederholung des ganzen für Quellen die weniger als 20 Mal erscheinen
to_drop = counts_df.loc[(counts_df['count'] < 7) & (counts_df['count'] > 1)]
to_drop.reindex()
to_drop
```

```python
# scheinbar nicht robuts, muss leider immer überprüft werden
to_drop = to_drop.drop([108, 110, 112, 118, 120, 122])
data = data[~data.publication.isin(to_drop.publication)]
# Datensatz ohne fehlerhafte Publikationen
data
```

## Schritt 2: Identifikation fehlerhaft importierter Artikel anhand der Artikelüberschrift

```python
# Identifikation der Einträge, welche keine Artikelüberschrift haben
no_text = pd.isnull(data['title'])
data[no_text]
```

```python
# Erstellen eines Datensatzes in dem alle Artikel eine Überschrift haben
data = data[~pd.isnull(data['title'])]
data
```

## Schritt 3: Vereinheitlichung des Datumsformats

```python
data.date.unique()
```

```python
# cleaning date column of wrongly parsed data
date_strings = data.date.value_counts()
date_strings = pd.DataFrame({'date_string':date_strings.index, 'count':date_strings.values})
date_strings.loc[date_strings['count'] == 1]
```

```python
to_drop = date_strings.loc[[272, 288, 294, 295, 297, 300, 304, 305, 308]]
#data[~data.date.isin(to_drop.date_string)]
data = data[~data.date.isin(to_drop.date_string)]
data.date.unique()
```

```python
# Datums-Strings in DateTime-Objekte konvertieren
import locale
locale.setlocale(locale.LC_ALL, "de_DE")
locale.getlocale()
```

```python
data['date_temp'] = pd.to_datetime(data['date'], format = "%d. %B %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%d.%B %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%-d. %B %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%-d.%B %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%d %B %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%d%B %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%-d %B %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%-d%B %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%d-%b-%y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%d-%b-%Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%d. %m. %Y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
data['date_temp'] = pd.to_datetime(data['date'], format = "%d. %m. %y", errors='coerce')
data.loc[data['date_temp'].notnull(), 'date'] = data['date_temp']
```
